## Demo: Understanding Spark Execution Plans Using explain()

```python
movies_df = spark.read.csv(
    "/Volumes/workspace/default/thinknyx_vol/movies.csv",
    header=True,
    inferSchema=True
)
display(movies_df)
```

```python
high_rated_df = movies_df.filter(
    (movies_df.averageRating > 9) & 
    (movies_df.numVotes > 1000000)
)
```

```python
high_rated_df.explain(True)
```
