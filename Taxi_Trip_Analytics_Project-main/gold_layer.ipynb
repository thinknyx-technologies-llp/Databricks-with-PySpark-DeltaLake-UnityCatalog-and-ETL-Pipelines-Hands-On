# Gold Layer — Business Metrics

```python
%sql
CREATE OR REPLACE TABLE taxicatalog.taxi_project_schema.gold_zone_revenue AS
SELECT
pickup_zone,
SUM(total_amount) AS total_revenue,
COUNT(*) AS total_trips
FROM taxicatalog.taxi_project_schema.silver_taxi_trips_enriched
GROUP BY pickup_zone
ORDER BY total_revenue DESC
```

```python
%sql
SELECT
pickup_zone,
pickup_latitude,
pickup_longitude,
SUM(total_amount) AS total_revenue
FROM taxicatalog.taxi_project_schema.silver_taxi_trips_enriched
GROUP BY
pickup_zone,
pickup_latitude,
pickup_longitude
LIMIT 10;
```

## 1. Peak Hours (Clean & Correct)

```python
%sql
CREATE OR REPLACE TABLE taxicatalog.taxi_project_schema.gold_peak_hours AS
SELECT
pickup_hour,
COUNT(*) AS total_trips,
SUM(total_amount) AS total_revenue
FROM taxicatalog.taxi_project_schema.silver_taxi_trips_enriched
GROUP BY pickup_hour
ORDER BY pickup_hour
```

```python
%sql

SELECT * FROM taxicatalog.taxi_project_schema.gold_peak_hours
```

## 2. Pickup vs Dropoff Time Analysis

```python
%sql
CREATE OR REPLACE TABLE taxicatalog.taxi_project_schema.gold_pickup_dropoff_time AS
SELECT
pickup_hour,
dropoff_hour,
COUNT(*) AS total_trips
FROM taxicatalog.taxi_project_schema.silver_taxi_trips_enriched
GROUP BY pickup_hour, dropoff_hour
```

## Revenue by Time of Day (Very Powerful)

```python
%sql
CREATE OR REPLACE TABLE taxicatalog.taxi_project_schema.gold_time_of_day_revenue AS
SELECT
CASE 
    WHEN pickup_hour BETWEEN 6 AND 11 THEN 'Morning'
    WHEN pickup_hour BETWEEN 12 AND 17 THEN 'Afternoon'
    WHEN pickup_hour BETWEEN 18 AND 22 THEN 'Evening'
    ELSE 'Night'
END AS time_of_day,
SUM(total_amount) AS total_revenue,
COUNT(*) AS total_trips
FROM taxicatalog.taxi_project_schema.silver_taxi_trips_enriched
GROUP BY time_of_day
```

## Revenue by Each Zone

```python
%sql
CREATE OR REPLACE TABLE taxicatalog.taxi_project_schema.gold_zone_revenue AS
SELECT
pickup_zone,
SUM(total_amount) AS total_revenue,
COUNT(*) AS total_trips
FROM taxicatalog.taxi_project_schema.silver_taxi_trips_enriched
GROUP BY pickup_zone
ORDER BY total_revenue DESC
```
