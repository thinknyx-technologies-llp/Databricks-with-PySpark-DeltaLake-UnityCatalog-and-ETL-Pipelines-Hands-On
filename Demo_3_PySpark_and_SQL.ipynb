```python
movies_df = spark.read.csv("/Volumes/workspace/default/thinknyx_vol/movies.csv", header=True, inferSchema=True)
```

```python
movies_df.display()
```

```python
movies_df.printSchema()
```

```python
movies_df.createOrReplaceTempView("movies")
```

```python
%sql
SELECT primaryTitle, averageRating, numVotes 
FROM movies
LIMIT 10;
```

```python
%sql
SELECT genres, ROUND(AVG(averageRating), 2) AS avg_rating
FROM movies
GROUP BY genres
ORDER BY avg_rating DESC
LIMIT 10;
```

```python
from pyspark.sql.functions import avg, round

movies_df.groupBy("genres") \
    .agg(round(avg("averageRating"), 2).alias("avg_rating")) \
    .orderBy("avg_rating", ascending=False) \
    .show(10)
```

```python
%sql

SELECT primaryTitle, averageRating
FROM movies
WHERE averageRating >= 7; 
```
