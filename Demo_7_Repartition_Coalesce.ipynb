
## Load DataSet

```python
movies_df = spark.read.csv(
    "/Volumes/workspace/default/thinknyx_vol/movies.csv",
    header=True,
    inferSchema=True
)

display(movies_df)
```

Check Number of Partitions Distribution

```python
# movies_df.rdd.getNumPartitions()

from pyspark.sql.functions import spark_partition_id

movies_with_pid = movies_df.withColumn(
    "partition_id",
    spark_partition_id()
)

display(movies_with_pid)
```

```python
movies_with_pid.select("partition_id").distinct().count()
```

## Repartition

```python
repartitioned_df = movies_df.repartition(5)
```

```python
repartitioned_with_pid = repartitioned_df.withColumn(
    "partition_id",
    spark_partition_id()
)

display(repartitioned_with_pid)
```

```python
repartitioned_with_pid.select("partition_id").distinct().count()
```

```python
repartitioned_df.explain(True)
```

## Coalesce

```python
coalesced_df = repartitioned_df.coalesce(2)
```

```python
coalesced_with_pid = coalesced_df.withColumn(
    "partition_id",
    spark_partition_id()
)

display(coalesced_with_pid)
```

```python
coalesced_with_pid.select("partition_id").distinct().count()
```

```python
coalesced_df.explain(True)
```

## Real Practical Scenario – Writing Files

```python
# Writing without Coalesce

repartitioned_df.write.mode("overwrite").csv("/Volumes/workspace/default/thinknyx_vol/movies_output_repartitioned")
```

```python
# Writing with Coalesce

coalesced_df.write.mode("overwrite").csv("/Volumes/workspace/default/thinknyx_vol/movies_output_coalesced")
```
