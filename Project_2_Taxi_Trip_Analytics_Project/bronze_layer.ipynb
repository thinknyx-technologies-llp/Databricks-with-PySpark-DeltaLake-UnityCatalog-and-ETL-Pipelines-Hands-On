# Bronze Layer - Raw Data Ingestion

```python
df_trips = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load("/Volumes/taxicatalog/taxi_project_schema/taxivolume/taxi_trip_data_raw.csv")

df_trips.show(5)
```

```python
df_trips.write.format("delta").saveAsTable("taxicatalog.taxi_project_schema.bronze_taxi_trips")
```
