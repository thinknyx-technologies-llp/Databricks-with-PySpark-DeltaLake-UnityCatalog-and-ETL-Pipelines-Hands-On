# Silver Layer

```python
%sql

CREATE OR REPLACE TABLE silver_walmart_sales
USING DELTA
AS
SELECT
  Store,
  TO_DATE(Date, 'yyyy-MM-dd') AS Date,
  Weekly_Sales,
  Holiday_Flag,
  Temperature,
  Fuel_Price,
  CPI,
  Unemployment,
  YEAR(TO_DATE(Date, 'yyyy-MM-dd')) AS Year,
  MONTH(TO_DATE(Date, 'yyyy-MM-dd')) AS Month
FROM workspace.default.walmart_dataset
WHERE Weekly_Sales IS NOT NULL
  AND Weekly_Sales > 0;
```

```python
%sql

SELECT * FROM silver_walmart_sales LIMIT 10;
```

```python
%sql

UPDATE silver_walmart_sales
SET 
    Weekly_Sales = ROUND(Weekly_Sales, 2),
    Temperature  = ROUND(Temperature, 2),
    Fuel_Price   = ROUND(Fuel_Price, 2),
    CPI          = ROUND(CPI, 2),
    Unemployment = ROUND(Unemployment, 2);
```
