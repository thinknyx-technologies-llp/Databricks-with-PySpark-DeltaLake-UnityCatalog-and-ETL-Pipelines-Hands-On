
# Gold Layer

```python
%sql

CREATE OR REPLACE TABLE gold_sales_by_store
USING DELTA
AS
SELECT
  Store,
  ROUND(SUM(Weekly_Sales), 2) AS Total_Sales
FROM workspace.default.silver_walmart_sales
GROUP BY Store;
```

```python
%sql
SELECT * FROM workspace.default.gold_sales_by_store
ORDER BY Total_Sales DESC;
```

```python
%sql
CREATE OR REPLACE TABLE gold_holiday_impact
USING DELTA
AS
SELECT
  Holiday_Flag,
  round(SUM(Weekly_Sales), 2) AS Total_Sales,
  round(AVG(Weekly_Sales), 2) AS Avg_Weekly_Sales
FROM silver_walmart_sales
GROUP BY Holiday_Flag;
```

```python
%sql
SELECT * FROM gold_holiday_impact;
```

```python
%sql
CREATE OR REPLACE TABLE gold_monthly_sales
USING DELTA
AS
SELECT
  Year,
  Month,
  round(SUM(Weekly_Sales), 2) AS Monthly_Sales
FROM silver_walmart_sales
GROUP BY Year, Month
ORDER BY Year, Month;
```

```python
%sql
SELECT * FROM gold_monthly_sales;
```

```python
%sql
SELECT 
  make_date(Year, Month, 1) AS Sales_Date,
  Monthly_Sales
FROM gold_monthly_sales
ORDER BY Sales_Date;
```
