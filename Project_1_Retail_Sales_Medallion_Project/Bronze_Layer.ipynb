```python
%sql
-- Ingestion
select * from csv.`s3://databricks-demo-thinknyx/`
with (header=true);
```

```python
%sql
CREATE TABLE workspace.default.walmart_dataset
USING DELTA
LOCATION 's3://databricks-demo-thinknyx/bronze/'
AS
SELECT
    *,
    current_timestamp() AS created_at,
    input_file_name() AS source_path,
    uuid() AS record_id
FROM csv.`s3://databricks-demo-thinknyx/`
WITH (header = true, inferSchema = true);
```

```python
%sql
ALTER TABLE workspace.default.walmart_dataset
ADD COLUMN temp_category STRING;
```

```python
%sql
UPDATE workspace.default.walmart_dataset
SET temp_category = 
    CASE  
        WHEN Temperature < 40 THEN 'Cold'  
        WHEN Temperature BETWEEN 40 AND 70 THEN 'Mild'  
        ELSE 'Hot'  
    END;
```

```python
%sql
SELECT * FROM workspace.default.walmart_dataset VERSION AS OF 3 WHERE Holiday_Flag = 1;
```
